# Average Week of Arrests in Baltimore in Chicago

```python
# Convert Daily Counts to Pandas DataFrames

DF_BPD_arrest_daily_count = pd.DataFrame(BPD_arrest_daily_count,columns=['count'])
DF_Chi_arrest_daily_count = pd.DataFrame(Chi_arrest_daily_count,columns=['count'])
```

```python
# Group by day of week

DF_BPD_arrest_daily_count['DayOfWeek'] = DF_BPD_arrest_daily_count.index.dayofweek
BPDgrouped = DF_BPD_arrest_daily_count.groupby('DayOfWeek')

DF_Chi_arrest_daily_count['DayOfWeek'] = DF_Chi_arrest_daily_count.index.dayofweek
Chigrouped = DF_Chi_arrest_daily_count.groupby('DayOfWeek')

# Average week (store Chi+Balt in "AvgWeek" DF)
AvgWeek = pd.merge(BPDgrouped.mean(),Chigrouped.mean(),how='outer',on=BPDgrouped.mean().index)
del AvgWeek['key_0']
AvgWeek.columns = ['Baltimore','Chicago']

# Calculate frequencies
AvgWeek['Baltimore Freq'] = AvgWeek['Baltimore'] / sum(AvgWeek['Baltimore'])
AvgWeek['Chicago Freq'] = AvgWeek['Chicago'] / sum(AvgWeek['Chicago'])
```

```python
# Plot typical weeks (average daily arrests)

Plot_AvgWeek = AvgWeek['Baltimore'].plot(style='.-')
Plot_AvgWeek = AvgWeek['Chicago'].plot(style='.-')

Plot_AvgWeek.set_ylim(bottom=0,top=300)
Plot_AvgWeek.set_ylabel('Average Daily Arrests')

# Create week labels
weekdays = 'Mon Tue Wed Thu Fri Sat Sun'.split()

# Change the xtick labels to be the weekdays
Plot_AvgWeek.set_xticklabels(weekdays)

Plot_AvgWeek.set_title('Average Daily Arrests in Baltimore and Chicago')
Plot_AvgWeek.legend(loc='lower left')
```

# Plot proportions of weekly arrests in Chicago

```python
Plot_AvgWeek_Freq = AvgWeek['Baltimore Freq'].plot(style='.-')
Plot_AvgWeek_Freq = AvgWeek['Chicago Freq'].plot(style='.-')

Plot_AvgWeek_Freq.set_ylim(bottom=0,top=0.2)
Plot_AvgWeek_Freq.set_ylabel('Percentage of Average Weekly Arrests')

# Create week labels
weekdays = 'Mon Tue Wed Thu Fri Sat Sun'.split()

# Change the xtick labels to be the weekdays
Plot_AvgWeek_Freq.set_xticklabels(weekdays)

Plot_AvgWeek_Freq.set_title('Daily Proportion of Average Weekly Arrests')
Plot_AvgWeek_Freq.legend(loc='lower left')
```

# Proportion of Violent Crimes ("Rape" "Murder" "Assault") by Day

```python
# Do less violent crimes occur on Sundays!?  Change in crime description prevalence...
# Need to test these for statistical significance

BPD_arrests['DayOfWeek'] = BPD_arrests.index.dayofweek

# Group arrests by day of week and T/F whether "Murder" or "Rape" or "Assault" appear in Charge Descrp.
grouped = BPD_arrests.groupby([BPD_arrests['DayOfWeek'], BPD_arrests['chargeDescription'].str.contains('Murder')|BPD_arrests['chargeDescription'].str.contains('Assault')|BPD_arrests['chargeDescription'].str.contains('Rape')]).describe()

BPD_Violent = pd.DataFrame(np.zeros((7,2)),columns=['Num_Violent','Prop_Violent'])
BPD_Violent.index = 'Mon Tues Wed Thurs Fri Sat Sun'.split()

# Proportion of violent crime arrests for day
BPD_Violent.loc['Mon','Prop_Violent'] = grouped.loc[(0,True,'count'),'arrest'] / ( grouped.loc[(0,True,'count'),'arrest']+grouped.loc[(0,False,'count'),'arrest'] )
BPD_Violent.loc['Tues','Prop_Violent'] = grouped.loc[(1,True,'count'),'arrest'] / ( grouped.loc[(1,True,'count'),'arrest']+grouped.loc[(1,False,'count'),'arrest'] )
BPD_Violent.loc['Wed','Prop_Violent'] = grouped.loc[(2,True,'count'),'arrest'] / ( grouped.loc[(2,True,'count'),'arrest']+grouped.loc[(2,False,'count'),'arrest'] )
BPD_Violent.loc['Thurs','Prop_Violent'] = grouped.loc[(3,True,'count'),'arrest'] / ( grouped.loc[(3,True,'count'),'arrest']+grouped.loc[(3,False,'count'),'arrest'] )
BPD_Violent.loc['Fri','Prop_Violent'] = grouped.loc[(4,True,'count'),'arrest'] / ( grouped.loc[(4,True,'count'),'arrest']+grouped.loc[(4,False,'count'),'arrest'] )
BPD_Violent.loc['Sat','Prop_Violent'] = grouped.loc[(5,True,'count'),'arrest'] / ( grouped.loc[(5,True,'count'),'arrest']+grouped.loc[(5,False,'count'),'arrest'] )
BPD_Violent.loc['Sun','Prop_Violent'] = grouped.loc[(6,True,'count'),'arrest'] / ( grouped.loc[(6,True,'count'),'arrest']+grouped.loc[(6,False,'count'),'arrest'] )

# Actual number of violent crime arrests
BPD_Violent.loc['Mon','Num_Violent'] = grouped.loc[(0,True,'count'),'arrest'] 
BPD_Violent.loc['Tues','Num_Violent'] = grouped.loc[(1,True,'count'),'arrest']
BPD_Violent.loc['Wed','Num_Violent'] = grouped.loc[(2,True,'count'),'arrest']
BPD_Violent.loc['Thurs','Num_Violent'] = grouped.loc[(3,True,'count'),'arrest']
BPD_Violent.loc['Fri','Num_Violent'] = grouped.loc[(4,True,'count'),'arrest']
BPD_Violent.loc['Sat','Num_Violent'] = grouped.loc[(5,True,'count'),'arrest']
BPD_Violent.loc['Sun','Num_Violent'] = grouped.loc[(6,True,'count'),'arrest']
```

```python
# Plot total number of violent crimes by day
BPD_Violent['Num_Violent'].plot(ylim=[600,900],title='Number of Violent Crime Arrests by Day')
```

```python
# Plot proportion of violent crimes out of total crimes by day
BPD_Violent['Prop_Violent'].plot(ylim=[0,0.3],title='Proportion of Violent Crime Arrests Each Day')
```
